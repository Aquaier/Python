```python
import time
import random
import matplotlib.pyplot as plt

class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
        self.size = 0

    def add_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next is not None:
                current = current.next
            current.next = new_node
        self.size += 1

    def remove_first(self):
        if self.head is None:
            return None
        else:
            removed_data = self.head.data
            self.head = self.head.next
            self.size -= 1
            return removed_data

class Queue:
    def __init__(self):
        self.linked_list = LinkedList()

    def enqueue(self, value):
        self.linked_list.add_end(value)

    def dequeue(self):
        return self.linked_list.remove_first()

    def is_empty(self):
        return self.linked_list.size == 0

    def peek(self):
        if self.is_empty():
            return 'Kolejka jest pusta'
        else:
            return self.linked_list.head.data

    def display(self):
        return self.linked_list

n_wartosci = [10, 50, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000]
enqueue_times = []

for n in n_wartosci:
    q = Queue()
    enqueue_time = 0
    for _ in range(n):
        item = random.randint(1, 1000)
        q.enqueue(item)
    average_enqueue_time = enqueue_time / n
    enqueue_times.append(average_enqueue_time)

plt.plot(n_wartosci, enqueue_times, label='Czas dodawania do kolejki')
plt.xlabel('Rozmiar danych')
plt.ylabel('Czas dodawania do kolejki (s)')
plt.legend()
plt.title('Wykres czasu dodawania do kolejki')
plt.grid()
plt.show()
```

```python
class Stack:
    def __init__(self):
        self.items = []

    def isEmpty(self):
        return self.items == []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        return self.items.pop()

    def peek(self):
        return self.items[len(self.items)-1]

    def size(self):
        return len(self.items)

    def print(self):
        return self.items

    def konwerter(self, value):
      stack = Stack()
      binarna = ''
      while value > 0:
        reszta = value % 2
        stack.push(reszta)
        value = value // 2
      while not stack.isEmpty():
        binarna = binarna + str(stack.pop())
      return binarna
liczba_dziesietna_test = 20
test = Stack()
binarna_wynik = test.konwerter(liczba_dziesietna_test)
print(f'dziesietna: {liczba_dziesietna_test}, binarna: {binarna_wynik}')
```

```python
class Node(object):
    def __init__(self, value):
        self.right = None
        self.left = None
        self.value = value

    def set_left(self, value):
        self.left = Node(value)

    def set_right(self, value):
        self.right = Node(value)

    def __str__(self):
        return str(self.value)


class BinTree(object):
    def __init__(self, value):
        self.root = Node(value)

    def h(self):
        return self._h(self.root)

    def _h(self, tree=None):
        if tree.left is None and tree.right is None:
            return 0
        if tree.left is None:
            return self._h(tree.right)
        if tree.right is None:
            return self._h(tree.left)
        return max(self._h(tree.left), self._h(tree.right)) + 1

    def preorder(self, tree=None):
        if tree is None:
            tree = self.root
        result = ''
        result += str(tree.value) + ' '
        if tree.left is not None:
            result += self.preorder(tree.left)
        if tree.right is not None:
            result += self.preorder(tree.right)
        return str(result)

    def inorder(self, tree=None):
        if tree is None:
            tree = self.root
        result = ''
        if tree.left is not None:
            result += self.inorder(tree.left)
        result += str(tree.value) + ' '
        if tree.right is not None:
            result += self.inorder(tree.right)
        return result

    def postorder(self, tree=None):
        if tree is None:
            tree = self.root
        result = ''
        if tree.left is not None:
            result += self.postorder(tree.left)
        if tree.right is not None:
            result += self.postorder(tree.right)
        result += str(tree.value) + ' '
        return result

    def leaf(self):
        return self._leaf(self.root)

    def _leaf(self, tree=None):
        leaf_counter = 0
        if tree.left is None and tree.right is None:
            return 1
        elif tree.left is None:
            leaf_counter += self._leaf(tree.right)
        elif tree.right is None:
            leaf_counter += self._leaf(tree.left)
        else:
            leaf_counter += self._leaf(tree.left) + self._leaf(tree.right)
        return leaf_counter

    def weight(self):
        return self._weight(self.root)

    def _weight(self, tree=None):
        weight_counter = 0
        if tree.left is None and tree.right is None:
            return 1
        elif tree.left is None:
            weight_counter += 1 + self._weight(tree.right)
        elif tree.right is None:
            weight_counter += 1 + self._weight(tree.left)
        else:
            weight_counter += 1 + self._weight(tree.left) + self._weight(tree.right)
        return weight_counter

    def avg(self):
        sum_values, count_nodes = self._average(self.root)
        if count_nodes == 0:
            return 0
        return sum_values / count_nodes

    def _average(self, tree=None):
        if tree is None:
            return 0, 0
        sum_left, count_left = self._average(tree.left)
        sum_right, count_right = self._average(tree.right)
        sum_values = tree.value + sum_left + sum_right
        count_nodes = 1 + count_left + count_right
        return sum_values, count_nodes


t = BinTree(5)
t.root.set_left(3)
t.root.set_right(7)
t.root.left.set_left(2)
t.root.left.set_right(4)
t.root.right.set_left(6)
t.root.right.set_right(8)
print(t.avg())
```

```python
  class BSTNode:
      def __init__(self, val=None):
          self.left = None
          self.right = None
          self.val = val

      def insert(self, val):
          if self.val is None:
              self.val = val
          else:
              if val <= self.val:
                  if self.left:
                      self.left.insert(val)
                  else:
                      self.left = BSTNode(val)
              else:
                  if self.right:
                      self.right.insert(val)
                  else:
                      self.right = BSTNode(val)

      def search(self, key):
          if self.val == key:
              return True
          if key < self.val:
              if not self.left:
                  return False
              return self.left.search(key)
          else:
              if not self.right:
                  return False
              return self.right.search(key)

      def get_min_iter(self):
          current = self
          while current.left is not None:
              current = current.left
          return current.val

      def get_min_rek(self):
          if self.left is None:
              return self.val
          return self.left.get_min_rek()

      def get_max_iter(self):
          current = self
          while current.right is not None:
              current = current.right
          return current.val

      def get_max_rek(self):
          if self.right is None:
              return self.val
          return self.right.get_max_rek()

      def preorder(self, vals=None):
          if vals is None:
              vals = []
          if self.val is not None:
              vals.append(self.val)
          if self.left is not None:
              self.left.preorder(vals)
          if self.right is not None:
              self.right.preorder(vals)
          return vals

      def inorder(self, vals=None):
          if vals is None:
              vals = []
          if self.left is not None:
              self.left.inorder(vals)
          vals.append(self.val)
          if self.right is not None:
              self.right.inorder(vals)
          return vals

      def find_k(self, k):
          vals = self.inorder()
          if 1 <= k <= len(vals):
              return vals[k - 1]
          else:
              return None

      def remove_node(self, key):
          if key < self.val:
              if self.left:
                  self.left = self.left.remove_node(key)
          elif key > self.val:
              if self.right:
                  self.right = self.right.remove_node(key)
          else:
              if not self.left:
                  return self.right
              elif not self.right:
                  return self.left
              min_val = self.right.get_min_iter()
              self.val = min_val
              self.right = self.right.remove_node(min_val)
          return self


      def najblizsza_wartosc(self, value):
          najblizsza = float()
          curr = self
          while curr:
              if abs(curr.val - value) < abs(najblizsza - value):
                  najblizsza = curr.val
              if value < curr.val:
                  curr = curr.left
              elif value > curr.val:
                  curr = curr.right
              else:
                  return curr.val
          return najblizsza



root = BSTNode(10)
root.insert(6)
root.insert(12)
root.insert(4)
root.insert(2)
root.insert(15)
root.insert(11)
liczba = 5
bliska_wartosc = root.najblizsza_wartosc(liczba)
print(f"Najbliższa wartosc: {liczba}, {bliska_wartosc}")


```
